```python
# where your file is located
FILE_LOCATION = '/content/drive/My Drive/pattern_realdata/pexam_pub_12.sas7bdat'
```

```python
df = pd.read_sas(FILE_LOCATION)
df
df.head()
```

```python
table = df[['IDind', 'U24B', 'U23', 'U36']]
table.head()
```

```python
# Define in which columns to look for missing values.
table1 = table.dropna(subset=['U24B', 'U23', 'U36'])
table1
```

```python
x = table1[['U23', 'U36']]
y = table1["U24B"]
x = sm.add_constant(x)
model = sm.OLS(y, x).fit()
y_pred = model.predict(x)

df = pd.DataFrame({"Actual": y, "Predicted": y_pred})

table_data_new = pd.DataFrame({ "x1"    : table1["U23"],
                                "x2"    : table1["U36"],
                                "y"     : df["Actual"],
                                "y_hat" : df["Predicted"]})
table_data_new["e1"] = table_data_new["y"] - table_data_new["y_hat"]
table_data_new["e2"] = table_data_new["y"] - table_data_new["y_hat"]
table_data_new
```

```python
table_data_new.plot(x ='x1', y='e1', kind = 'scatter')
table_data_new.plot(x ='x2', y='e2', kind = 'scatter')
```

```python
data1 = table_data_new.loc[table_data_new["x1"] > 0]
data1 = data1.loc[data1["e1"] >= -60]
data1
```

```python
data1.plot(x ='x1', y='e1', kind = 'scatter')
```

```python
data2 = table_data_new.loc[table_data_new["x2"] > 0]
data2 = data2.loc[data2["e2"] >= -50]
data2 = data2.loc[data2["e2"] < 75]
data2
```

```python
data2.plot(x ='x2', y='e2', kind = 'scatter')
```

```python
data1.plot(x ='x1', y='e1', kind = 'scatter')
data2.plot(x ='x2', y='e2', kind = 'scatter')
```
